```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
```

```python
df = pd.read_csv("datasets/processed_data_with_stages.csv")
```

```python
df.head()
```

### Correlation

```python
plt.figure(figsize=(12,8))


ax1=plt.subplot(2, 2, 1)
sns.set_style("darkgrid")
plot_acf(df['DAC'].values,lags=50,ax=ax1, color="#c6474b", vlines_kwargs={"colors": "#c6474b"})
plt.xlabel('')
plt.ylabel("")
sns.despine()
plt.title("DAC (Algoritmo de Disciplinado)")


ax2=plt.subplot(2, 2, 2)

sns.set_style("darkgrid")
plot_acf(df['Phase Error'].values,lags=50,ax=ax2, color="#c6474b", vlines_kwargs={"colors": "#c6474b"})
plt.xlabel('')
plt.ylabel("")
plt.title("Error de Fase (Phase Error)")
sns.despine()

ax3=plt.subplot(2, 2, 3)

sns.set_style("darkgrid")
plot_acf(df['Temperature'].values,lags=50,ax=ax3, color="#c6474b", vlines_kwargs={"colors": "#c6474b"})
plt.xlabel('')
plt.ylabel("")
plt.title("Temperatura")
sns.despine()

ax4=plt.subplot(2, 2, 4)

sns.set_style("darkgrid")
plot_acf(df['Frequency Error'].values,lags=50,ax=ax4, color="#c6474b", vlines_kwargs={"colors": "#c6474b"})
plt.xlabel('')
plt.ylabel("")
plt.title("Error de Frecuencia (Frequency Error)")
sns.despine()

plt.suptitle("Autocorrelación", fontsize=18)
plt.tight_layout()

#plt.savefig("fotos/autocorrelacion_4variables.png")
```

```python
plt.figure(figsize=(14,4))


ax1=plt.subplot(1, 3, 1)
sns.set_style("darkgrid")
sns.lineplot(data=pd.Series(sm.tsa.stattools.ccf(df['DAC'].values, df['Frequency Error'].values, adjusted=False))[:200], color="#c6474b")
plt.xlabel('Lag')
plt.ylabel("Correlación")
sns.despine()
plt.title("DAC - Frequency Error")


ax2=plt.subplot(1, 3, 2)

sns.set_style("darkgrid")
sns.lineplot(data=pd.Series(sm.tsa.stattools.ccf(df['DAC'].values, df['Phase Error'].values, adjusted=False))[:200], color="#c6474b")
plt.xlabel('Lag')
plt.ylabel("")
plt.title("DAC - Phase Error")
sns.despine()

ax3=plt.subplot(1, 3, 3)

sns.set_style("darkgrid")
sns.lineplot(data=pd.Series(sm.tsa.stattools.ccf(df['Frequency Error'].values, df['Phase Error'].values, adjusted=False))[:200], color="#c6474b")
plt.xlabel('Lag')
plt.ylabel("")
plt.title("Frequency Error - Phase Error")
sns.despine()

plt.suptitle("Correlación Cruzada (Cross-Correlation)", fontsize=18)
plt.tight_layout()

#plt.savefig("fotos/cross_correlation.png")
```
